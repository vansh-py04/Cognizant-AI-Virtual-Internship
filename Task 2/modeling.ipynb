```python
# converting timestamp from object to datetime in each dataframe 

def convert_timestamp(data,column):
    dummy = data.copy()
    dummy[column] = pd.to_datetime(dummy[column], format="%Y-%m-%d %H:%M:%S")
    return dummy
```

```python
sales_df = convert_timestamp(sales_df,"timestamp")
temp_df = convert_timestamp(temp_df,"timestamp")
stock_df = convert_timestamp(stock_df,"timestamp")
```

## Merge Data

The datasets can be merged on the common timestamp column, but we need the final data on an hourly basis.
So we will convert the timestamp column to hour

```python
from datetime import datetime

def convert_ts_to_hourly(data,column):
    dummy = data.copy()
    new_ts = dummy[column].tolist()
    new_ts = [i.strftime("%Y-%m-%d %H:00:00") for i in new_ts]
    new_ts = [datetime.strptime(i,"%Y-%m-%d %H:00:00") for i in new_ts]
    dummy[column] = new_ts
    return dummy
```

```python
sales_df = convert_ts_to_hourly(sales_df,"timestamp")
sales_df.head()
```

```python
stock_df = convert_ts_to_hourly(stock_df,"timestamp")
temp_df = convert_ts_to_hourly(temp_df,"timestamp")
```

Aggregate datasets based on time and product_id

```python
sales_agg = sales_df.groupby(["timestamp","product_id"]).agg({"quantity":"sum"}).reset_index()
sales_agg.head()
```

```python
stock_agg = stock_df.groupby(["timestamp","product_id"]).agg({"estimated_stock_pct":"mean"}).reset_index()
stock_agg.head()
```

```python
temp_agg = temp_df.groupby(["timestamp"]).agg({"temperature":"mean"}).reset_index()
temp_agg.head()
```

Now, we are ready to merge our data. We will use the `stock_agg` table as our base table, and we will merge our other 2 tables onto this.

```python
merged_df = stock_agg.merge(sales_agg,on=["timestamp","product_id"], how="left")
merged_df.head()
```

```python
merged_df = merged_df.merge(temp_agg, on=["timestamp"],how="left")
merged_df.head()
```

We can see that `quantity` column has some null values for products which did not have any sales for a particular hour. We will deal with this by filling the null values with 0.

```python
merged_df["quantity"] = merged_df["quantity"].fillna(0)
merged_df.info()
```

```python
product_category = sales_df[["product_id","category"]]
product_category = product_category.drop_duplicates()
product_category
```

```python
product_price = sales_df[["product_id","unit_price"]]
product_price = product_price.drop_duplicates()
product_price
```

Merging these columns as well.

```python
merged_df = merged_df.merge(product_category, on=["product_id"],how="left")
merged_df = merged_df.merge(product_price, on=["product_id"],how="left")
```

```python
merged_df.head()
```

## Feature Engineering

First we will convert timetsamp to day_of_the_month, day_of_the_week, hour.

```python
merged_df["day_of_the_month"] = merged_df["timestamp"].dt.day
merged_df["day_of_the_week"] = merged_df["timestamp"].dt.dayofweek
merged_df["hour"] = merged_df["timestamp"].dt.hour
merged_df.drop(columns=['timestamp'], inplace=True)
merged_df.head()
```

We can perform specific one hot encoding on the category column, and then drop the product_id column.

```python
merged_df = pd.get_dummies(merged_df, columns=['category'],dtype=float)
merged_df.head()
```

```python
merged_df = merged_df.drop(columns=["product_id"])
merged_df.head()
```

```python
merged_df.info()
```

## Modelling

```python
y = merged_df["estimated_stock_pct"]
X = merged_df.drop(columns=["estimated_stock_pct"])
```

```python
X.shape
```

```python
y.shape
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.preprocessing import StandardScaler
```

```python
model = RandomForestRegressor(random_state=42)
scaler = StandardScaler()
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,y,random_state=42)
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
model.fit(x_train,y_train)
```

```python
pred = model.predict(x_test)
error = mean_absolute_error(y_test,pred)
print(f"MAE of Model:{error:.3f}")
```

---
## Feature Importance

```python
import matplotlib.pyplot as plt

features = [i.split("__")[0] for i in X.columns]
importances = model.feature_importances_
indices = np.argsort(importances)

fig, ax = plt.subplots(figsize=(10, 20))
plt.title('Feature Importances')
plt.barh(range(len(indices)), importances[indices], color='g', align='center')
plt.yticks(range(len(indices)), [features[i] for i in indices])
plt.xlabel('Relative Importance')
plt.show()
```

This feature importance visualisation tells us:

- The product categories were not that important
- The unit price and temperature were important in predicting stock
- The hour of day was also important for predicting stock

```python
feature_importance = pd.DataFrame({'Feature': features, 'Importance': importances}).sort_values(by='Importance', ascending=False)
```

```python
plt.figure(figsize=(10, 7))
colors = plt.get_cmap('Set2').colors  # Use a colormap for better colors
explode = [0.1 if i == 0 else 0 for i in range(6)]  # Explode the first slice

plt.pie(
    feature_importance["Importance"][:6],
    labels=feature_importance["Feature"][:6],
    shadow=True,
    autopct='%1.1f%%',  # Display percentages
    startangle=140,  # Start angle for better orientation
    colors=colors,  # Apply colors
    explode=explode,  # Apply explosion effect
    wedgeprops={'edgecolor': 'black'}  # Add black edge to wedges for better separation
)

plt.title('Top 6 Feature Importances')
plt.show()
```
